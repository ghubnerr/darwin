```python
from os import path
from glob import glob
import ffmpeg

PATH = "../src/assets/atari/*.gif"

files = tuple(path.abspath(f) for f in glob(PATH))
```

```python
from os import environ

p = environ.get("PATH")
print(p)
```

```python
for f in files:
    filename = path.basename(f)
    stream = ffmpeg.input(f)
    stream = ffmpeg.output(stream, f"tmp/img/{filename}.png", **{"vframes": 1})
    ffmpeg.run(stream, cmd="/opt/homebrew/bin/ffmpeg")
```

```python
from zipfile import ZipFile
import ffmpeg
import os


# filename = path.basename(files[0]).split(".")[0]
filename = "zaxxon"
stream = ffmpeg.input("/Users/luis/dev/initBuild-ai1/src/assets/atari/zaxxon.gif")
stream = ffmpeg.output(stream, "tmp/img/out_%d.png")
ffmpeg.run(stream, cmd="/opt/homebrew/bin/ffmpeg")

imgs = glob("tmp/img/*.png")
imgs = sorted(imgs, key=lambda x: int(x.split("_")[1].split(".")[0]))
imgs = tuple(map(os.path.abspath, imgs))

with ZipFile(f"tmp/{filename}.zip", "w") as f:
    for img in imgs:
        f.write(img, arcname=path.basename(img))

# make_archive(filename, "zip", "tmp/img")

for f in imgs:
    os.remove(f)
```
