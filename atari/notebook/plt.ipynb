```python
import seaborn as sns
import pandas as pd
from itertools import count, chain, repeat
import itertools
```

```python
def round_robin(*iterables):
    "roundrobin('ABC', 'D', 'EF') --> A D E B F C"
    # Recipe credited to George Sakkis
    [x for x in itertools.chain.from_iterable(itertools.zip_longest(*iterables)) if x]
```

```python
rewards = list(range(1, 11_000))
losses = list(range(-10_000, 0))

data = []
for i, (r, l) in enumerate(zip(rewards, losses)):
    data += [(i, "reward", r)]
    data += [(i, "loss", l)]


# data = pd.DataFrame({"rewards": rewards, "losses": losses})
data = pd.DataFrame(
    data=data,
    columns=["epoch", "type", "value"],
)
# data = data.transpose()
data.head()
```
