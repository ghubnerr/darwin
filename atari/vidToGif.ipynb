```python
from os import path
from glob import glob
import ffmpeg

FOLDER = "./assets/riverraid/videos"
PATH = f"{FOLDER}/*.mp4"

files = tuple(path.abspath(f) for f in glob(PATH))
```

```python
GIF_FOLDER = "./assets/riverraid/gif"
for f in files:
    filename = path.basename(f)
    stream = ffmpeg.input(f)
    stream = ffmpeg.output(
        stream,
        f"{GIF_FOLDER}/{filename.split('.')[0]}.gif",
        **{"loop": 0, "f": "gif"},
    )
    ffmpeg.run(stream, cmd="/opt/homebrew/bin/ffmpeg")
```
